```python
import warnings
from pathlib import Path
import pandas as pd
```

```python
# File Path
food_access_data_to_load = Path("../Data/food_access.csv")

# Read shcool metrics into Pandas DataFrames
food_access_df = pd.read_csv(food_access_data_to_load)

food_access_df.head()
```

```python
food_access_df.info()
```

```python
print(food_access_df.columns)
```

```python
# Reorder the Columns
column_order = ['State', 'County', 'Population', 'Housing Data.Total Housing Units', 'Housing Data.Residing in Group Quarters', 
                'Vehicle Access.1/2 Mile', 'Vehicle Access.1 Mile', 'Vehicle Access.10 Miles', 'Vehicle Access.20 Miles', 
                'Low Access Numbers.Children.1/2 Mile', 'Low Access Numbers.Children.1 Mile', 'Low Access Numbers.Children.10 Miles', 'Low Access Numbers.Children.20 Miles', 
                'Low Access Numbers.Low Income People.1/2 Mile', 'Low Access Numbers.Low Income People.1 Mile', 'Low Access Numbers.Low Income People.10 Miles', 'Low Access Numbers.Low Income People.20 Miles', 
                'Low Access Numbers.People.1/2 Mile', 'Low Access Numbers.People.1 Mile', 'Low Access Numbers.People.10 Miles', 'Low Access Numbers.People.20 Miles', 
                'Low Access Numbers.Seniors.1/2 Mile', 'Low Access Numbers.Seniors.1 Mile', 'Low Access Numbers.Seniors.10 Miles', 'Low Access Numbers.Seniors.20 Miles'
               ]

food_access_df = food_access_df[column_order]
```

```python
# Rename the Columns
food_access_df.rename(columns={'Housing Data.Total Housing Units': 'Total_Housing_Units', 
                                 'Housing Data.Residing in Group Quarters': 'Residing_in_Group_Quarters', 
                                 'Vehicle Access.1/2 Mile': 'No_Vehicle_Half_Mile', 
                                 'Vehicle Access.1 Mile': 'No_Vehicle_1_Mile', 
                                 'Vehicle Access.10 Miles': 'No_Vehicle_10_Miles', 
                                 'Vehicle Access.20 Miles': 'No_Vehicle_20_Miles', 
                                 'Low Access Numbers.Children.1/2 Mile': 'Children_Half_Mile', 
                                 'Low Access Numbers.Children.1 Mile': 'Children_1_Mile', 
                                 'Low Access Numbers.Children.10 Miles': 'Children_10_Miles', 
                                 'Low Access Numbers.Children.20 Miles': 'Children_20_Miles', 
                                 'Low Access Numbers.Low Income People.1/2 Mile': 'Low_Income_Half_Mile', 
                                 'Low Access Numbers.Low Income People.1 Mile': 'Low_Income_1_Mile', 
                                 'Low Access Numbers.Low Income People.10 Miles': 'Low_Income_10_Miles', 
                                 'Low Access Numbers.Low Income People.20 Miles': 'Low_Income_20_Miles', 
                                 'Low Access Numbers.People.1/2 Mile': 'Population_Half_Mile', 
                                 'Low Access Numbers.People.1 Mile': 'Population_1_Mile', 
                                 'Low Access Numbers.People.10 Miles': 'Population_10_Miles', 
                                 'Low Access Numbers.People.20 Miles': 'Population_20_Miles', 
                                 'Low Access Numbers.Seniors.1/2 Mile': 'Seniors_Half_Mile', 
                                 'Low Access Numbers.Seniors.1 Mile': 'Seniors_1_Mile', 
                                 'Low Access Numbers.Seniors.10 Miles': 'Seniors_10_Miles', 
                                 'Low Access Numbers.Seniors.20 Miles': 'Seniors_20_Miles' 
                                }, 
                        inplace=True
                       )
```

```python
print(food_access_df.columns)
```

```python
# Change column values to percent of population
percentage_columns = ['Residing_in_Group_Quarters', 
                      'No_Vehicle_Half_Mile', 'No_Vehicle_1_Mile', 'No_Vehicle_10_Miles', 'No_Vehicle_20_Miles', 
                      'Children_Half_Mile', 'Children_1_Mile', 'Children_10_Miles', 'Children_20_Miles', 
                      'Low_Income_Half_Mile', 'Low_Income_1_Mile', 'Low_Income_10_Miles', 'Low_Income_20_Miles', 
                      'Population_Half_Mile', 'Population_1_Mile', 'Population_10_Miles', 'Population_20_Miles', 
                      'Seniors_Half_Mile', 'Seniors_1_Mile', 'Seniors_10_Miles', 'Seniors_20_Miles'
                     ]

# Convert to percentage of the population
for col in percentage_columns:
    food_access_df[col] = (food_access_df[col] / food_access_df['Population']) * 100

food_access_df.head()
```

```python
food_access_df.info()
```

```python
# File Path
county_key_data_to_load = Path("../Outputs/county_key.csv")

# Read shcool metrics into Pandas DataFrames
county_key_df = pd.read_csv(county_key_data_to_load)

county_key_df.head()
```
